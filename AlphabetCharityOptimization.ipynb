```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
app_df = application_df.drop(columns=['EIN', 'NAME'])
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `app_replace`
app_values = app_df['APPLICATION_TYPE'].value_counts()
app_replace = app_values[app_values < 10].index.tolist()

# Replace in dataframe
for app in app_replace:
    app_df['APPLICATION_TYPE'] = app_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
app_df['APPLICATION_TYPE'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
value = app_df['CLASSIFICATION'].value_counts()
filter_value = value[value > 1]
print(filter_value)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
class_values = app_df['CLASSIFICATION'].value_counts()
class_replace = class_values[class_values < 10].index.tolist()

# Replace in dataframe
for cls in class_replace:
    app_df['CLASSIFICATION'] = app_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
app_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
app_numeric_df = pd.get_dummies(app_df, columns=['INCOME_AMT', 'APPLICATION_TYPE', 'CLASSIFICATION'])
```

```python
app_numeric_df.head()
```

```python
from scipy.sparse import random
# Split our preprocessed data into our features and target arrays
y = app_numeric_df['IS_SUCCESSFUL']
x = app_numeric_df.drop(columns =['AFFILIATION', 'USE_CASE', 'ORGANIZATION', 'STATUS', 'SPECIAL_CONSIDERATIONS', 'IS_SUCCESSFUL'])

# Split the preprocessed data into a training and testing dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=88)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(x_train)

# Scale the data
X_train_scaled = X_scaler.transform(x_train)
X_test_scaled = X_scaler.transform(x_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

nn2 = tf.keras.models.Sequential()

# First hidden layer
nn2.add(tf.keras.layers.Dense(units=80, activation="relu", input_dim=53))

# Second hidden layer
nn2.add(tf.keras.layers.Dense(units=30, activation="relu", input_dim=53))

# Output layer
nn2.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn2.summary()
```

```python
nn2.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn2.fit(X_train_scaled, y_train, epochs=100)
```
